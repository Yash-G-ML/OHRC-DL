```python
source = 'E:\Yash\subsamples\subsamples_ohrc_new'
clahe = cv2.createCLAHE(tileGridSize=(8,8))

for file in os.listdir(source):
    strip = np.load(os.path.join(source,file))
    
    gt_placeholder = []
    for i in range(strip.shape[0]):
        cl = clahe.apply(strip[i])
        gt_placeholder.append(cl)
    gt = np.array(gt_placeholder)
    np.save('E:\Yash\subsamples\subsamples_ohrc_new_gt'+'\GT_'+file,gt)
```
